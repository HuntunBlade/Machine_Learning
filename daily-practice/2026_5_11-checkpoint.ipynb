```python
import numpy as np
```

```python
arr = np.arange(1,26).reshape(5,5)
print("原始矩阵：\n",arr)
```

```python
# 1. 取单个元素（第2行第3列，注意索引从0开始！）
print("\n第2行第3列:", arr[1, 2])

# 2. 取一整行（第1行）
print("\n第1行:", arr[0, :])

# 3. 取一整列（第2列）
print("\n第2列:", arr[:, 1])

# 4. 取一个子矩阵（中间2行2列）
print("\n中间2x2子矩阵:\n", arr[1:3, 1:3])

# 5. 花式索引：取第0, 2, 4行，这很实用
print("\n第0,2,4行:\n", arr[[0, 2, 4], :])
```

```python
# 接上面的arr
# 找出所有大于10的元素的位置
mask = arr > 10
print("哪些元素>10？\n", mask)

# 直接取出所有大于10的元素
print("\n所有>10的值:", arr[arr > 10])

# 练习：把arr中所有>10的变成-1，其余不变
arr_copy = arr.copy()  # 好习惯，不破坏原始数据
arr_copy[arr_copy > 10] = -1
print("\n>10全变-1:\n", arr_copy)
```
