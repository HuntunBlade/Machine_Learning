```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 年龄缺失？用同客舱等级的中位数来填
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))

print("数据大小（行，列）：", df.shape)
print("\n前3行长这样：")
print(df.head(3))
```
