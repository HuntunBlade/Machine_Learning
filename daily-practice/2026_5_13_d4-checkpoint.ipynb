```python
import pandas as pd
import numpy as np

# 加载数据（记得改路径）
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 1. 先看看年龄缺失了多少
print("年龄缺失人数:", df['Age'].isnull().sum())

# 2. 最简单的填充：用中位数填满
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

# 3. 验证一下：现在还有缺失吗？
print("填充后缺失人数:", df['Age'].isnull().sum())

# 4. 进阶填充（选做）：按“乘客等级”分组填充中位数
#    头等舱的乘客可能更年长，分组填充更精确
#df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
df['Age'] = df['Age'].fillna(median_age)
```

```python
# 1. 不同性别的存活率
print("===== 性别 vs 存活 =====")
print(df.groupby('Sex')['Survived'].mean())

# 2. 不同船舱等级的存活率
print("\n===== 船舱 vs 存活 =====")
print(df.groupby('Pclass')['Survived'].mean())

# 3. 有兄弟姐妹的乘客存活率更高吗？
print("\n===== 兄弟姐妹数 vs 存活 =====")
print(df.groupby('SibSp')['Survived'].mean())
```

```python
import matplotlib.pyplot as plt

# 设置中文显示（如果遇到乱码，可以先用英文标签）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号

# 图1：乘客年龄分布直方图
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)  # 1行2列，第1个图
plt.hist(df['Age'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# 图2：不同性别存活率柱状图
plt.subplot(1, 2, 2)  # 1行2列，第2个图
survival_by_sex = df.groupby('Sex')['Survived'].mean()
plt.bar(survival_by_sex.index, survival_by_sex.values, color=['blue', 'pink'])
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')

plt.tight_layout()  # 自动调整子图间距
plt.show()
```
