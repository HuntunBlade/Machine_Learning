```python
import pandas as pd
```

```python
# 加载泰坦尼克数据集（一个非常经典的公共数据）
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
# 1. 看前5行，感受一下长什么样
print("===== 前5行 =====")
print(df.head())
```

```python
# 2. 看整体信息：多少行？哪些列有缺失？每列是什么类型？
print("\n===== 数据信息 =====")
df.info()
```

```python
# 3. 看数字列的统计摘要：平均年龄、票价最高最低...
print("\n===== 统计摘要 =====")
print(df.describe())
```

```python
# 4. 看某个单一列（用 df['列名'] ）
print("\n===== 年龄列前5个值 =====")
print(df['Age'].head())
```

```python
# 5. 从表格里“切”出一小块：取前10行的 ['Name', 'Survived'] 列
print("\n===== 前10人的名字与存活 =====")
print(df.loc[:9, ['Name', 'Survived']])
```
