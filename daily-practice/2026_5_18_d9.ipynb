```python
import pandas as pd
import numpy as np
url = 'https://gitee.com/py233/datasets/raw/master/titanic.csv'
df = pd.read_csv(url)
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
print("缺失值搞定，还剩：", df.isnull().sum().sum(), "个")
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
print("✅ 文字全部转成数字")
print(df[['Sex', 'Embarked']].head())
```

```python
# 选 6 个最经典的特征
feature_cols = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize']

X = df[feature_cols].copy()
y = df['Survived'].copy()

print("特征形状：", X.shape)  # (891, 6)
print("标签形状：", y.shape)  # (891,)
print("\n特征预览：")
print(X.head())
```

```python
# 存成 csv，明天直接读
df_ready = X.copy()
df_ready['Survived'] = y
df_ready.to_csv('titanic_ready.csv', index=False)
print("✅ 已保存 titanic_ready.csv，明天直接训练！")
```
