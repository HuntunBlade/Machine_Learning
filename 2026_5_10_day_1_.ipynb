```python
my_array = np.array([1, 2, 3, 4, 5])
print(my_array)
print("形状:", my_array.shape)  # 几行几列？
print("维度:", my_array.ndim)  # 是几维的？
print("数据类型:", my_array.dtype)
print("全零:", np.zeros(5))
print("全一:", np.ones((2, 3)))  # 2行3列
print("等差数列:", np.arange(0, 10, 2))  # 从0到10，步长为2
print("随机数:", np.random.rand(3, 2))  # 3行2列的0到1之间随机数
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("加法:", a + b)
print("点乘:", a * b)       # 对应位置相乘
print("点积(内积):", np.dot(a, b))  # 1*4 + 2*5 + 3*6 = 32
```
